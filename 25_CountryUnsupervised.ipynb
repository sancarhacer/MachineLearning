```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("25-country_data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import math

def plot_all_histograms(df, title_prefix=""):
    num_cols = df.select_dtypes(include=[np.number]).columns
    n_cols = 3
    n_rows = math.ceil(len(num_cols) / n_cols)

    plt.figure(figsize=(5 * n_cols, 4 * n_rows))

    for i, col in enumerate(num_cols, 1):
        plt.subplot(n_rows, n_cols, i)
        sns.histplot(df[col], kde=True, bins=30)
        plt.title(f"{title_prefix}{col}")
        plt.xlabel("")
        plt.ylabel("")

    plt.tight_layout()
    plt.show()
plot_all_histograms(df)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
df2 = df.drop("country", axis=1)
df2.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2 = scaler.fit_transform(df2)
df.columns
```

```python
df2 = pd.DataFrame(df2, columns=['child_mort', 'exports', 'health', 'imports', 'income',
       'inflation', 'life_expec', 'total_fer', 'gdpp'])
plot_all_histograms(df2)
```

```python
from sklearn.decomposition import PCA
pca = PCA() #n_component:None
#Bu durumda PCA, verideki tm olas bileenleri retir (yani hibir boyut indirgeme yapmaz, sadece dntrr).yani verini farkl bir koordinat sistemine tar.
pca_df2 = pd.DataFrame(pca.fit_transform(df2))
pca_df2
```

```python
pca.explained_variance_ratio_
```

```python
plt.step(list(range(1,10)), np.cumsum(pca.explained_variance_ratio_))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.ylabel("Variance Covered")
plt.title("Variance Covered")
plt.show()
```

```python
pca_df2 = pca_df2.drop(columns = [3,4,5,6,7,8])
pca_df2
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pca_df2)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
model = KMeans(n_clusters=3)
model.fit(pca_df2)
```

```python
labels = model.labels_
silhouette_score(pca_df2,labels)
```

```python
df['Class'] = labels
df
```

```python
fig, ax = plt.subplots(nrows = 1, ncols= 2 , figsize=(15,5))

plt.subplot(1,2,1)
sns.boxplot(data=df, x="Class", y="child_mort")
plt.title("child_mort vs class")

plt.subplot(1,2,2)
sns.boxplot(data=df, x="Class", y="income")
plt.title("income vs class")

plt.show()
```

```python
# 0 -> no budget needed
# 1 -> budget needed
# 2 -> in between
import plotly.express as px
pca_df2.insert(0, column= "Country", value = df['country'])
pca_df2['Class'] = labels
pca_df2.loc[pca_df2['Class'] == 1, 'Class'] = "No Budget Needed"
pca_df2.loc[pca_df2['Class'] == 2, 'Class'] = "In Between"
pca_df2.loc[pca_df2['Class'] == 0, 'Class'] = "Budget Needed"
```

```python
pca_df2.head()
```

```python
fig = px.choropleth(
    pca_df2[['Country', 'Class']],
    locationmode = "country names",
    locations = "Country",
    title = "Needed Budget by Country",
    color = pca_df2['Class'],
    color_discrete_map= {
                        "Budget Needed" : "Red",
                        "In Between" : "Yellow",
                        "No Budget Needed": "Green"
    })
fig.update_geos(fitbounds = "locations", visible = True)
fig.show()
```
