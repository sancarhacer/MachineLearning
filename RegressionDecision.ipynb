In this section, we are using the LazyPredict library to quickly train and compare multiple regression models on the diabetes dataset.


```python
#https://github.com/shankarpandala/lazypredict
!pip install lazypredict
```

```python
from lazypredict.Supervised import LazyRegressor
from sklearn import datasets
from sklearn.utils import shuffle
import numpy as np
diabetes  = datasets.load_diabetes()
type(diabetes)
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
X, y = shuffle(diabetes.data, diabetes.target, random_state=13)
X = X.astype(np.float32)
offset = int(X.shape[0] * 0.9)

X_train, y_train = X[:offset], y[:offset]
X_test, y_test = X[offset:], y[offset:]
y_train
```

```python
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train, y_test)
models
```

The goal of this code is to quickly benchmark multiple regression algorithms on the diabetes dataset without writing separate code for each model. This allows us to identify the best-performing algorithms efficiently.
