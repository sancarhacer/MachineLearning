```python
df = df.drop(["Unnamed: 0"], axis=1)
df.describe()
```

```python
df = df.drop(df[df["x"]==0].index)
df = df.drop(df[df["y"]==0].index)
df = df.drop(df[df["z"]==0].index)
df.describe()
```

```python
df = df[(df["depth"]<75)&(df["depth"]>45)]
df = df[(df["table"]<80)&(df["table"]>40)]
df = df[(df["y"]<30)]
df = df[(df["z"]<30)&(df["z"]>2)]
df.describe()
```

```python
sns.scatterplot(x=df["z"], y=df["price"])
plt.show()
```

```python
df['clarity'].value_counts()
```

```python
X= df.drop(["price"],axis =1)
y= df["price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25, random_state=15)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoders = {}
for col in ['cut', 'color', 'clarity']:
    encoders[col] = LabelEncoder()
    X_train[col] = encoders[col].fit_transform(X_train[col])
    X_test[col] = encoders[col].transform(X_test[col])
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.svm import SVR
```

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf','linear']}
grid = GridSearchCV(SVR(), param_grid, refit = True,verbose = 3,n_jobs=-1) # n_jobs=-1 means use all processor
grid.fit(X_train_scaled, y_train)
```

```python
## grid.best_params_
```

```python
svr=SVR(C=1000, gamma=0.1, kernel='rbf')
```

```python
from sklearn.metrics import r2_score
svr.fit(X_train_scaled, y_train)
y_pred=svr.predict(X_test_scaled)
score=r2_score(y_test,y_pred)
print("R2 Score", score)
```

```python
import pickle
with open('26-diamond_model_complete.pkl', 'wb') as f:
    pickle.dump({
        'model': svr,
        'encoders': encoders,
        'scaler': scaler
    }, f)
```
